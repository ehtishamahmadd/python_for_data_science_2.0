# 03- How to rename columns?

```python
# adding more vales in the columns and then checking the dataframe
df = pd.DataFrame({'A col': [1, 2, 3,6,7,8,9,22], 'B col': [4, 5, 6,11 ,22,12,13,34]})
df.head(10)
```

```python
# rename columns
df.rename(columns={'A col': 'col_a', 'B col': 'col_b'}, inplace=True) 
# inplace=True: change the original dataframe
df.head()
```

```python
# other ways of renaming columns
df.columns=['col_aa', 'col_bb']
df
```

```python
# to replace any character or string with another
df.columns= df.columns.str.replace('_', ' ')
df
```

```python
# again replacing any character or string with another

df.columns= df.columns.str.replace(' ', '_')
df
```

```python
# adding prefix to columns
df = df.add_prefix('baba_')
df
```

```python
#  adding suffix to columns
df = df.add_suffix('baba_')
df
```

```python
# again renaming the columns
df.columns = {'col_a','col_b'}
df
```

# 12- Split dataframe into two subsets

```python
# dataset downlaod from the internet
import pandas as pd # importing the libraries for datasets
import seaborn as sns # importing the libraries for plotting
df= sns.load_dataset('titanic')
df.head()
```

```python
# for splitting we need to look into the length of the dataset
len(df)
```

```python
# for splitting we need to look into the shape of the dataset also
df.shape
```

```python
# splitting the dataset
kashti_1 = df.sample(frac=0.50, random_state=1) # taking 0.5(50%) fraction of the dataset
kashti_1.shape
```

```python
# now again splitting the dataset by removing first dataset from the main dataset, so getting the remaining data
kashti_2 = df.drop(kashti_1.index)
kashti_2.shape
```

```python
# now checking the data using head command
kashti_1.head()
# it will give random data from the dataset
```

```python
# now checking the data using head command
kashti_2.head()
# it will give random data from the dataset
```

# 14- Filtering a datasets

```python
# checking the data using head command
df.head()
```

```python
# seeing the unique values in the dataset(sex column)
df.sex.unique()
```

```python
# taking only data of female in the sex column
df[(df.sex == 'female')]
```

```python
# taking only data of embark_town
df.embark_town.unique()
```

```python
# if we want just the 'Southampton town' data
df[(df.embark_town=='Southampton')].shape
```

```python
# if we want just females who were from Southampton town, we will use two filters
df[(df.embark_town=='Southampton') &
    (df.sex == 'female')]
```

```python
# if we want just females who were from Southampton town or Queenstown
df[((df.embark_town=='Southampton') |
    (df.embark_town=='Queenstown')) &
    (df.sex == 'female')]
```

```python
# filtering the only people from Queenstown and Southampton
df[(df.embark_town.isin(['Queenstown','Southampton']))].head()
```

```python
# filtering age less than 18
df[df.age < 18].shape
```

```python
# filter the data by age more than 30 and various other filters
df[(df.age > 30) & 
    ((df.embark_town == 'Southampton') |
    (df.embark_town == 'Queenstown')) &
    (df.pclass == 1) &
    ((df.embarked !='Q') |
    (df.embarked !='S')) &
    (df.fare < 30)].shape
```
