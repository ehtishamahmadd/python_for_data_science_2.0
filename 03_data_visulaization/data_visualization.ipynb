```python
# making plots in matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# loading data  
df = sns.load_dataset('tips')   # tips is a dataset in seaborn
df.head()
```

```python
sns.lineplot(x='day', y='total_bill', data=df)  # line plot
```

```python
sns.barplot(x='day', y='total_bill', data=df) # barplot is a bar chart
```

```python
sns.boxenplot(x='day', y='total_bill', data=df) # boxenplot is a boxplot
```

```python
sns.boxplot(x='day', y='total_bill', data=df) # boxplot is a boxplot
```

```python
sns.catplot(x='day', y='total_bill', data=df) # categorical plot
```

```python
sns.distplot(df['total_bill']) # distribution plot
```

```python
sns.lmplot(x='total_bill', y='tip', data=df) # linear model plot
```

```python
sns.heatmap(df.corr(), annot=True) # heatmap
```

```python
sns.pairplot(df) # pairplot is a scatterplot with regression lines
```

```python
sns.factorplot(x='day', y='total_bill', data=df) # factorplot is a categorical plot
```
